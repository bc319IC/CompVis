## 2. Edge detection.

Perform edge detection using Prewitt filtering, as well as Gaussian + Prewitt filtering.

### 2.1 Implement 3x3 Prewitt filters and convolve with the noisy image. (10 points)

```python
# Design the Prewitt filters
prewitt_x = np.array([[1,0,-1],
                     [1,0,-1],
                     [1,0,-1]])
prewitt_y = np.array([[1,1,1],
                     [0,0,0],
                     [-1,-1,-1]])

# Prewitt filtering
pfi_x = scipy.signal.convolve2d(image_noisy,prewitt_x)
pfi_y = scipy.signal.convolve2d(image_noisy,prewitt_y)

# Calculate the gradient magnitude
grad_mag = np.sqrt(pfi_x**2 + pfi_y**2)

# Print the filters (provided)
print('prewitt_x:')
print(prewitt_x)
print('prewitt_y:')
print(prewitt_y)

# Display the gradient magnitude image (provided)
plt.imshow(grad_mag, cmap='gray')
plt.gcf().set_size_inches(10, 8)
```

### 2.2 Implement a function that generates a 2D Gaussian filter given the parameter $\sigma$. (10 points)

```python
# Design the Gaussian filter
def gaussian_filter_2d(sigma):
    # sigma: the parameter sigma in the Gaussian kernel (unit: pixel)
    #
    # return: a 2D array for the Gaussian kernel
    h = np.full([6*sigma+1,6*sigma+1],0.0)
    for x in range(-3*sigma,3*sigma+1):
        for y in range(-3*sigma,3*sigma+1):
            h[x+3*sigma][y+3*sigma] = np.exp(-(x**2+y**2)/(2*sigma**2))/(2*np.pi*sigma**2)
    return h

# Visualise the Gaussian filter when sigma = 3 pixel (provided)
sigma = 3
h = gaussian_filter_2d(sigma)
plt.imshow(h)
```

### 2.3 Perform Gaussian smoothing ($\sigma$ = 3 pixels), followed by Prewitt filtering, show the gradient magnitude image. (5 points)

```python
# Perform Gaussian smoothing before Prewitt filtering
gf = gaussian_filter_2d(3)
image_smoothed = scipy.signal.convolve2d(image_noisy,gf)

# Prewitt filtering
pfi_gx = scipy.signal.convolve2d(image_smoothed,prewitt_x)
pfi_gy = scipy.signal.convolve2d(image_smoothed,prewitt_y)

# Calculate the gradient magnitude
grad_mag = np.sqrt(pfi_gx**2 + pfi_gy**2)

# Display the gradient magnitude image (provided)
plt.imshow(grad_mag, cmap='gray', vmin=0, vmax=100)
plt.gcf().set_size_inches(10, 8)
```

### 2.5 Implement a function that generates a 1D Gaussian filter given the parameter $\sigma$. Generate 1D Gaussian filters along x-axis and y-axis respectively. (10 points)

```python
# Design the Gaussian filter
def gaussian_filter_1d(sigma):
    # sigma: the parameter sigma in the Gaussian kernel (unit: pixel)
    #
    # return: a 1D array for the Gaussian kernel
    h = np.full([6*sigma+1,1],0.0)
    for x in range(-3*sigma,3*sigma+1):
        h[x+3*sigma][0] = np.exp(-x**2/(2*sigma**2))/(np.sqrt(2*np.pi)*sigma)
    return h

# sigma = 7 pixel (provided)
sigma = 7

# The Gaussian filter along x-axis. Its shape is (1, sz).
h_x = gaussian_filter_1d(sigma).T

# The Gaussian filter along y-axis. Its shape is (sz, 1).
h_y = gaussian_filter_1d(sigma)

# Visualise the filters (provided)
plt.subplot(1, 2, 1)
plt.imshow(h_x)
plt.subplot(1, 2, 2)
plt.imshow(h_y)
```

### 2.7 Comment on the Gaussian + Prewitt filtering results and the computational time. (9 points)

The computational complexity of the non-separable filter is O(M*N*K^2) as there is K^2 multiplications and K^2-1 summations at each pixel. As there are M*N pixels that means overall there are M*N*K^2 multiplications and M*N(K^2-1) summations which gives the previously mentioned complexity, where the image is M by N and the kernel size is K by K.
For the separable filter the computational complexity is O(M*N*K) as now we have K multiplications and K-1 summations at each pixel but this process has to be repeated twice, giving 2M*N*K multiplications and 2M*N(K-1) summations.
For both the above cases, the image size is the same so the difference in computational complexity is of factor K. The larger the K, the larger the difference. In this case, the sigma value is 7 which means K = 7*3*2+1 = 43 so the kernel is 43 by 43 meaning the separable filter should be considerably faster than the non-separable which is evident from the previous tasks.

The [Conv2D](https://pytorch.org/docs/stable/nn.html#conv2d) operator takes an input array of dimension NxC1xXxY, applies the filter and outputs an array of dimension NxC2xXxY. Here, since we only have one image with one colour channel, we will set N=1, C1=1 and C2=1. You can read the documentation of Conv2D for more detail.

```python
# Import libaries (provided)
import torch
```

### 3.1 Expand the dimension of the noisy image into 1x1xXxY and convert it to a Pytorch tensor. (7 points)

### 3.2 Create a Pytorch Conv2D filter, set its kernel to be a 2D Gaussian filter. (7 points)

```python
# A 2D Gaussian filter when sigma = 3 pixel (provided)
sigma = 3
h = gaussian_filter_2d(sigma)
h = np.expand_dims(h, axis=0)
h = np.expand_dims(h, axis=0)
print(h.shape)

# Create the Conv2D filter (provided)
conv = torch.nn.Conv2d(1, 1, kernel_size=3, padding=1, bias=False)

# Set the kernel weight
conv.weight = torch.nn.Parameter(torch.from_numpy(h))
```
